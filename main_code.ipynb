## Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id
 
Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
#write your query
query6 = '''select N.Country, N.League_name,Match.*
            from Match left join (select *,Country.name as Country, League.name as 'League_name' 
                                    from Country inner join League on Country.id = League.country_id   
            ) as N
            ON Match.country_id = N.country_id
            '''
result6 = pd.read_sql(query6,conn)
result6
```

## Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
#write your query
query7 = '''select League.name as 'Tên giải đấu', count(Match.league_id) as 'Số trận đấu'
            from Match inner join League ON Match.league_id = League.id
            Group by Match.league_id
            Order by count(Match.league_id) DESC
            '''
result7 = pd.read_sql(query7,conn)
result7
```

## Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
#write your query
query12 = ''' select T.team_long_name as 'Tên đội bóng', sum(Total) as 'Tổng bàn thắng'
            from
            (select country_id,league_id,home_team_api_id, sum(home_team_goal) as 'Total'
                from Match
                group by home_team_api_id
                union
                select country_id,league_id,away_team_api_id, sum(away_team_goal) as 'Bàn thắng sân khách'
                from Match
                group by away_team_api_id) as N 
            left join Team as T on N.home_team_api_id = T.team_api_id
            group by N.home_team_api_id
            order by Total DESC
            '''

result12 = pd.read_sql(query12,conn)
result12
```

##  Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
#write your query
query13 = '''select T.team_long_name as 'Tên đội bóng', sum(home_match) as 'Tổng số trận'
            from (select Team.team_long_name , count(Match.home_team_api_id) as 'home_match'
                from Match left join Team ON Match.home_team_api_id = Team.team_api_id
                Group by Match.home_team_api_id
             UNION ALL
                select Team.team_long_name, count(Match.away_team_api_id) as 'away_match'
                from Match left join Team ON Match.away_team_api_id = Team.team_api_id
                Group by Match.away_team_api_id) as T
            group by T.team_long_name
            order by sum(home_match) DESC
            '''
result13 = pd.read_sql(query13,conn)
result13
```

## Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
#write your query
query14_1 = ''' select T.Team_name as ' Tên đội', count(Result) as 'Số trận thắng'
            from (select Team.team_long_name as 'Team_name',M.home_team_goal,M.away_team_goal ,
                    (CASE 
                    when M.home_team_goal > M.away_team_goal then 'thắng'
                    when M.home_team_goal < M.away_team_goal then 'thua'
                    when M.home_team_goal = M.away_team_goal then 'hòa' END) as Result
                    from Match as M left join Team ON M.home_team_api_id = Team.team_api_id) as T
            where T.Result = 'thắng'
            group by T.Team_name
            order by count(Result) DESC
            '''
result14_1 = pd.read_sql(query14_1,conn)
result14_1
```

```python
#write your query
query14_2 = ''' select T.Team_name as ' Tên đội', count(Result) as 'Số trận thua'
            from (select Team.team_long_name as 'Team_name',M.home_team_goal,M.away_team_goal ,
                    (CASE 
                    when M.home_team_goal > M.away_team_goal then 'thắng'
                    when M.home_team_goal < M.away_team_goal then 'thua'
                    when M.home_team_goal = M.away_team_goal then 'hòa' END) as Result
                    from Match as M left join Team ON M.home_team_api_id = Team.team_api_id) as T
            where T.Result = 'thua'
            group by T.Team_name
            order by count(Result) ASC
            '''
result14_2 = pd.read_sql(query14_2,conn)
result14_2
```

```python
#write your query
query14_3 = ''' select T.Team_name as ' Tên đội', count(Result) as 'Số trận hòa'
            from (select Team.team_long_name as 'Team_name',M.home_team_goal,M.away_team_goal ,
                    (CASE 
                    when M.home_team_goal > M.away_team_goal then 'thắng'
                    when M.home_team_goal < M.away_team_goal then 'thua'
                    when M.home_team_goal = M.away_team_goal then 'hòa' END) as Result
                    from Match as M left join Team ON M.home_team_api_id = Team.team_api_id) as T
            where T.Result = 'hòa'
            group by T.Team_name
            order by count(Result) DESC
            '''
result14_3 = pd.read_sql(query14_3,conn)
result14_3
```

```python
result14 = [result14_1,result14_2,result14_3]
result14 = pd.merge(result14_1,result14_2,on=result14_2.columns[0])
result14 = pd.merge(result14,result14_3,on=result14_2.columns[0])
result14
```
